```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from tkinter import *

root = Tk()
root.config(background="light green")

label = Label(text="Pemograman Simulasi Regresi Linear", font="Normal 30", background='Light green')
button = Button(text="Tampilkan Dataset", font="Normal 25", activebackground='green')
button1 = Button(text="Tampilkan Scatter Plot", font="Normal 25", activebackground='green')
button2 = Button(text="Tampilkan Model Regresi", font="Normal 25", activebackground='green')
button3 = Button(text="Tampilkan Grafik", font="Normal 25", activebackground='green')

label.pack() 
button.pack()
button1.pack()
button2.pack()
button3.pack()
root.mainloop()




data = pd.read_csv('dataset.csv', delimiter=";")
data.head() 


X = data['Tahun'].values
Y = data['Mahasiswa'].values

plt.scatter(X, Y, label='Data')

plt.xlabel('Tahun')
plt.ylabel('Mahasiswa')

plt.legend()
plt.show()

mean_x = np.mean(X.copy())
mean_y = np.mean(Y.copy())

n = len(X)

num = 0
den = 0

for i in range(n):
    num += (X[i] - mean_x) * (Y[i] - mean_y)
    den += (X[i] - mean_x) ** 2

aX = float(num) / float(den)
b = mean_y - (aX * mean_x)

print('Mean Tahun = ' +str(mean_x))
print('Mean Mahasiswa = ' +str(mean_y))
print(aX,b)

max_x = np.max(X) + 10
min_x = np.min(X) - 10

x = np.linspace(min_x, max_x, 100)
y = aX + b * x

plt.plot(x, y, label='Linier Regression')
plt.scatter(X,Y, label='Scatter Plot')

plt.xlabel('Tahun')
plt.ylabel('Mahasiswa')

plt.legend()
plt.show()

y = aX + b * 2
print(y)



```
